```python
#show me the portion of the data containing the maximum high price and the minimum low price
```

```python
import pandas_datareader.data as web
import datetime as dt
import pandas as pd

ibm = web.DataReader('IBM', 'yahoo', dt.date.today()-dt.timedelta(days=30))
max_high = ibm.loc[ibm.High == ibm.High.max()]
min_low = ibm.loc[ibm.Low == ibm.Low.min()]
ibm_high_low = pd.concat((max_high, min_low))
print(ibm_high_low)
```

```python
#now plot a graph of the opening and closing prices for the 30-day period
```

```python
import pandas_datareader.data as web
import datetime as dt
import pylab as plt

ibm = web.DataReader('IBM', 'yahoo', dt.date.today()-dt.timedelta(days=30))

fig, ax = plt.subplots()
ax.plot(ibm.index, ibm.Open, ibm.Close)
```

```python
import pandas_datareader.data as web
import datetime as dt
import pylab as plt

ibm = web.DataReader('IBM', 'yahoo', dt.date.today()-dt.timedelta(days=30))

fig, ax = plt.subplots()
ax.plot(ibm.index, ibm.Open, ibm.Close)
ax.tick_params('x', labelrotation=45)
```
